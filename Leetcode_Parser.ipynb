```python
import os
import selenium
from selenium import webdriver
import time
from PIL import Image
import io
import requests
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from parsel import Selector
import time
# driver = webdriver.Chrome(ChromeDriverManager().install())
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
import csv

f = open('leetcode.csv','w',newline='')
fp = open('leetcode_problems.csv','w',newline='')
csv_writer = csv.writer(f)
csv_writer_p = csv.writer(fp)

# driver = webdriver.Chrome(ChromeDriverManager().install())
# driver.implicitly_wait(5)

#storing the problems so that if any duplicate occurs we can detect them
problems = []

#list all the companies csv files
# files = os.listdir()
in_dir = "./"
files = [os.path.join(pth, f) for pth, dirs, files in os.walk(in_dir) for f in files]

for csvfile in files:

    #ignore .py files and dataset files
    if csvfile.endswith('.py'):
        continue
    elif csvfile == 'leetcode.csv' or csvfile == 'leetcode_problems.csv':
        continue
    
    cur_file = open(csvfile)
    csv_cur_file_reader = csv.reader(cur_file)

    #getting the company name
    cur_comp = csvfile.split('_')[0]
    
    for cur_prob in csv_cur_file_reader:

        #problem name, problem statement, problem tags
        row = []

        #problem name, problem company
        row_fp = []

        #getting problem name
        cur_prob_name = cur_prob[1]
        if cur_prob_name in problems:
          continue

        try:

            #getting problem link
            cur_prob_link = cur_prob[5]

            #storing the pairs of problem names and companies
            row_fp.append(cur_prob_name)
            row_fp.append(cur_comp)
            csv_writer_p.writerow(row_fp)

            row.append(cur_prob_name)

            driver.get(cur_prob_link)

            #wait till the problem div is loaded
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'css-1eusa4c-TabContent')))
            
            prob = driver.find_elements(By.CLASS_NAME,'css-1eusa4c-TabContent')
            print(len(prob))
            prob = prob[0]

            #getting all the paras of the problem statement
            paras = prob.find_elements(By.TAG_NAME,'p')

            problem_statement = ''
            for para in paras:
                stmt = str(para.text.encode('UTF-8'))[2:-1]
                if stmt == "Example 1:":
                    break
                else:
                    problem_statement += stmt
            #print(problem_statement)
            row.append(problem_statement)
            # row.append(cur_comp)

            #getting all the tags (if present => hence try/catch)
            show_topics = driver.find_element(By.CLASS_NAME,'header__2RZv')
            show_topics.click()
            tags = driver.find_elements(By.CLASS_NAME,'tag__2PqS')
        
            for tag in tags:
                tag_text = tag.text
                row.append(tag_text)
            print(cur_prob_name,cur_comp)
            problems.append(cur_prob_name)

        except Exception as e:
            print(e)
        finally:
            #storing the problem name,statement and tags
            csv_writer.writerow(row)

fp.close()
f.close()


```
